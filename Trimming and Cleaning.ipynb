```python
import pandas as pd
import numpy as np
file_path = "/Users/darrelljarrett/Explainable Machine Learning for Diabetes Prediction/trimmed_2015_raw.csv"

use_cols = [
    "GENHLTH","BPHIGH4","CHOLCHK","TOLDHI2","DIABETE3",
    "SEX","EDUCA","INCOME2","SMOKE100","EXERANY2","_AGEG5YR","_BMI5"
]

df = pd.read_csv(file_path, usecols=use_cols, low_memory=False)
print("Loaded:", df.shape)
df.head()
```

```python
# Replace common BRFSS special codes with NaN
special_codes = [7, 9, 77, 99, 777, 999, 7777, 9999]
df = df.replace(special_codes, np.nan)

# Convert BMI back to proper scale (div by 100)
df["_BMI5"] = df["_BMI5"] / 100.0

# Optional: drop impossible BMI (sanity bounds)
df = df[(df["_BMI5"].isna()) | ((df["_BMI5"] >= 12) & (df["_BMI5"] <= 70))]

print("After special-code replacement:", df.shape)
df.isna().sum().sort_values(ascending=False).head(12)
```

```python
from collections import defaultdict
import numpy as np

cat_cols = [
    "GENHLTH","BPHIGH4","CHOLCHK","TOLDHI2",
    "SEX","EDUCA","INCOME2","SMOKE100","EXERANY2","_AGEG5YR"
]
num_cols = ["_BMI5"]

# Fill categorical with mode
modes = {c: df[c].mode(dropna=True)[0] if df[c].notna().any() else np.nan for c in cat_cols}
for c in cat_cols:
    df[c] = df[c].fillna(modes[c])

# Fill BMI with median
df["_BMI5"] = df["_BMI5"].fillna(df["_BMI5"].median())

# Drop rows with missing target (DIABETE3)
df = df.dropna(subset=["DIABETE3"])

print("After imputation:")
print(df.isna().sum())
print("\nShape:", df.shape)
```
